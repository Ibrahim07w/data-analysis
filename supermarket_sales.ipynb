```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/ibrah/Downloads/supermarket_sales.csv')
df.head()
```

```python
# to see if there are nulls in the data but there is no null
df.info()
```

```python
# to  get mean, std, min, and quartiles
df.describe()
```

```python
# to see the duplicates but there is no duplicates
df[df.duplicated() == True]
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
df['Time']
```

```python
df['Branch'].value_counts()
df['Gender'].value_counts()
df['Product line'].value_counts()
```

```python
df.groupby('Branch')['Total'].sum()
df.groupby('Product line')['Total'].sum()
```

```python
df[['Invoice ID','Total','Branch']].sort_values(by='Total', ascending=False).head(5)
```

```python
df.info()
```

```python
df['day of the week'] = df['Date'].dt.day_name()
df['month'] = df['Date'].dt.month
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
```

```python
# multiple filters
df[(df['Gender'] == 'Female') & (df['Product line'] == 'Health and beauty') & (df['Total'] > 600)].sort_values(by='Total', ascending=False)
```

```python
# Monthly Revenue per Branch
revenu = df.groupby(['Branch','month'])['Total'].sum().unstack()
revenu.plot(marker='o')
```

```python
#  Average Daily Gross Income
avg_gross = df.groupby('Date')['gross income'].mean()
avg_gross.plot(figsize=(10,8))
```

```python
# Sales by Payment Method
payment_count = df['Payment'].value_counts()
plt.tight_layout()
payment_count.plot(kind='pie', autopct='%1.1f%%', title='payment distribution')
```

```python
# Top 5 Product Lines by Total Sales
product_total = df.groupby('Product line')['Total'].sum().nlargest(5)
product_total.plot(kind='barh', title='top 5 products', color='skyblue')
plt.xlabel('total sales by product')
plt.xlim(50000, product_total.max() * 1.07)
```

```python
# Total Sales by Gender
gender_sales = df.groupby('Gender')['Total'].sum()
gender_sales.plot(kind='bar', color=['pink', 'blue'], title="Total Sales by Gender")
```

```python
# Hourly Sales Trend
hourly_sales = df.groupby('Hour')['Total'].mean()
hourly_sales.plot(marker='o', title="Avg Sales by Hour")
plt.xlabel("Hour")
plt.ylabel("Average Sales")
```

```python
# City-wise Revenue & Quantity
city_stats = df.groupby('City')[['Total', 'Quantity']].sum()
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
city_stats['Total'].plot(kind='bar', ax=ax[0], title="Revenue by City", color='green')
ax[0].set_ylabel("Revenue")
ax[0].set_ylim(100000, city_stats['Total'].max() * 1.1)
city_stats['Quantity'].plot(kind='bar', ax=ax[1], title="Quantity by City", color='orange')
ax[1].set_ylabel("Quantity")
ax[1].set_ylim(1500, city_stats['Quantity'].max() * 1.1)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(20,16), ncols=3, nrows=2)
plt.tight_layout()

# Calculate Monthly Sales
monthly_sales = df.groupby('month')['Total'].sum()
axes[0,0].bar(monthly_sales.index, monthly_sales.values, color='orange', edgecolor='yellow')
axes[0,0].set_title('monthly_sales')
axes[0,0].set_ylabel('revenue')
axes[0,0].set_xlabel('month')
axes[0,0].set_ylim(80000, 120000)

# average rating per product line
rating_by_product = df.groupby('Product line')['Rating'].mean()
axes[0,1].pie(rating_by_product.values, labels = rating_by_product.index,  autopct='%1.1f%%', startangle=90)
axes[0,1].set_title('rating by product')

# Quantity Sold by Customer Type
qty_customer_type = df.groupby('Customer type')['Quantity'].sum()
axes[0,2].barh(qty_customer_type.index, qty_customer_type.values, color='teal')
axes[0,2].set_xlim(2500,3000)
axes[0,2].set_title('quantity by customer type')

# female rating
female_ratings = df[df['Gender'] == 'Female']['Rating']
axes[1,1].hist(female_ratings, bins=10, color='salmon', edgecolor='plum')
axes[1,1].set_title('female rating')
```

```python
# sales in branch in hour
import seaborn as sns
hourly_branch_sales = df.groupby(['Branch','Hour'])['Total'].sum().unstack()
hourly_branch_sales.T.plot(figsize=(8,6))
plt.ylabel('sales')
plt.xlabel('hour')
plt.legend(loc=0)
plt.grid(True)
```

```python
sns.boxplot(x='Branch', y='Total', data=df, palette='Set2')
plt.title('Outlier Transactions in Total Sales by Branch', fontsize=14)
plt.xlabel('Branch', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
```

```python
# correlation matrix
numeric_cols = df.select_dtypes(include='number')
correlation = numeric_cols.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='plasma')
plt.title('Correlation Matrix')
```

```python
# Average Rating by City and Product Line
pivot = df.pivot_table(index='City', columns='Product line', values='Rating', aggfunc='mean')
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, cmap='Greens')
plt.title('Average Product Ratings by City and Product Line')
plt.show()
```
